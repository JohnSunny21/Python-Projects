```python
import ast 

# this module is used to evaluate the string representation of the list of dictionaries and convert them back to the list of dictionaries 
# for the next processing.
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.isnull().sum() # this step is used to check whether the film data conatians any empty values for any columns
# Here we can see the overview has 3 empty values so we need to drop them or delete the 3 film with 3 null overview values
```

```python
movies.dropna(inplace=True)

#dropna(): This method removes rows (by default) that contain missing values (NaNs).
#inplace=True: This modifies the original DataFrame. The changes are applied directly, 
#and no new DataFrame is created.inplace=False (default behavior): This does not modify the 
#original DataFrame. Instead, it returns a new DataFrame with the rows or columns removed, 
#while the original DataFrame remains unchanged."""
```

```python
movies.shape
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
# This is an example to demonstrate the ast module in which the above string is converted to respectivce data.
```

```python
# this method is used to get only three cast members names converted into list so that the main leads will be in the list.
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter += 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
# or you can use another method 
movies['cast'] = movies['cast'].apply(lambda x: x[0:3]) # first convert to list and then slicing it using list slicing 
# with lambda functionde
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5) # this method is used to randomly get some data from the films.
```

```python
# This method is used to remove all the spaces betweeen the words like James Cameron to JamesCameron
# To avoid ambiguity James Cameron tag and James tag both are different and should be a different tag
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
    
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split()) # to split all the words in the overview as it is a string.
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns= ['overview','genres','keywords','cast','crew'])
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x : " ".join(x))
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: x.lower()) # making all the tags lowercase
new.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
"""
Stemming is a process in natural language processing (NLP) that reduces words to their root or base form. The idea is to remove suffixes
and prefixes so that related words map to a common root. This helps in reducing the complexity of text data and making analysis more efficient.
Example
Words: "running", "runner", "ran"
Stem: "run"
Even though the words "running", "runner", and "ran" have different endings and usages, stemming reduces them all to "run"."""
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
new['tags'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key= lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    movies_container = sorted(list(enumerate(similarity[index])),reverse=True,key= lambda x: x[1])[1:6]
    # or use 
    #distances = sorted(list(enumerate(similarity[index])),reverse=True,key= lambda x: x[1])
    #for i in distances[1:6]:
    for i in movies_container:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
