```python
import ast 

# this module is used to evaluate the string representation of the list of dictionaries and convert them back to the list of dictionaries 
# for the next processing.
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.isnull().sum() # this step is used to check whether the film data conatians any empty values for any columns
# Here we can see the overview has 3 empty values so we need to drop them or delete the 3 film with 3 null overview values
```

```python
movies.dropna(inplace=True)

#dropna(): This method removes rows (by default) that contain missing values (NaNs).
#inplace=True: This modifies the original DataFrame. The changes are applied directly, 
#and no new DataFrame is created.inplace=False (default behavior): This does not modify the 
#original DataFrame. Instead, it returns a new DataFrame with the rows or columns removed, 
#while the original DataFrame remains unchanged."""
```

```python
movies.shape
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# this method is used to get only three cast members names converted into list so that the main leads will be in the list.
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter += 1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
# or you can use another method 
movies['cast'] = movies['cast'].apply(lambda x: x[0:3]) # first convert to list and then slicing it using list slicing 
# with lambda functionde
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5) # this method is used to randomly get some data from the films.
```

```python
# This method is used to remove all the spaces betweeen the words like James Cameron to JamesCameron
# To avoid ambiguity James Cameron tag and James tag both are different and should be a different tag
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
    
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split()) # to split all the words in the overview as it is a string.
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```
